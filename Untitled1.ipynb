```python
import pandas as pd
```

```python
df = pd.read_excel('data_all.xlsx')
```

```python
df.columns
```

```python
df = df.drop('Geburtsdatum', axis='columns')
```

```python
colreasonsel1 = df['Pat.Nr.'].apply(lambda x: isinstance(x, str))
```

```python
colreason1 = df.loc[colreasonsel1].rename({'Pat.Nr.': 'reason'}, axis='columns')
```

```python
colreason1
```

```python
colreason2 = df.loc[df['Name'].apply(lambda x: isinstance(x, str) and not ',' in x and not 'Dr.' in x)].rename({'Name': 'reason'}, axis='columns')
```

```python
colreason2
```

```python
colreason = pd.concat([colreason1, colreason2])
```

```python
colreason
```
